One of the most common magic commands is the `%matplotlib inline` '
This command will enable plots to be displayed directly inside the notebook

````python
%matplotlib inline
````

````python
"""Simple demo of scatter plot"""
import numpy as np
import matplotlib.pyplot as plt

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
# Calculates the area
area = np.pi * (15 * np.random.rand(n))**2
#plot dta
plt.scatter(x,y,s=area, c=colors, alpha = 0.5)
plt.show()
````

#### Python list vs Numpy arrays
- elements of different types `vs` same types
- variable size nesting is possible `vs` same size
- no operations at list level `vs` operations allowed at the array level
- Uses a lot of memory `vs` less memory
