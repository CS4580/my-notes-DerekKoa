```python
import pandas as pd

df_hotel = pd.read_csv('../hotel_bookings.csv')
df_hotel
```

```python
df_hotel.drop('company', inplace=True,axis=1) #inplace fixes the original
```

```python
df_hotel.head(5)
```

```python
#filling in missing records in children with 0
df_hotel.children = df_hotel.children.fillna(0)
```

```python
df_hotel = df_hotel.dropna(subset=['country'])
```

```python
df_hotel
```

## duplicates

- Practice how to handle duplicate values

```python
#total # of duplicates of dataset
df_hotel.duplicated().sum()
```

```python
#extract the dataframe containing the duplicates
df_hotel.loc[df_hotel.duplicated(), :]
```

```python
#drop duplicated data
df_hotel.drop_duplicates(inplace=True,keep='first')
```

## different data types

### datetime

```python
# convert the column "object" to a given type, in this case, datetime
df_hotel['reservation_status_date'] = pd.to_datetime(df_hotel['reservation_status_date'])
```

```python
#verify it
df_hotel.info()
```

```python
# create the year, month and day 

df_hotel['reservation_status_year'] = df_hotel['reservation_status_date'].dt.year
df_hotel['reservation_status_month'] = df_hotel['reservation_status_date'].dt.month
df_hotel['reservation_status_day'] = df_hotel['reservation_status_date'].dt.day
df_hotel.head()
```

```python
#remove drop temporary date time
df_hotel.drop('reservation_status_date', inplace=True, axis = 1)
```

## bad or illogical records or values

```python
# check how many recs have 0 for adults
(df_hotel['adults'] == 0).sum()

#drop any records whose'adult' is zero
df_hotel.drop(df_hotel[df_hotel['adults'] == 0].index, inplace=True)
#uses boolean index to filter the data
```
